#### Horizontal and Vertical lines

```python
plt.figure(figsize=(18,10))
plt.scatter(df['avg'], df['strike_rate'], s=df['runs'])

plt.axhline(130, color='red')
plt.axvline(30, color='green')

for i in range(df.shape[0]):
  plt.text(df['avg'].values[i], df['strike_rate'].values[i], df['batter'].values[i])
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True, figsize=(10,6))

ax[0].scatter(df['avg'], df['strike_rate'], color='green')
ax[1].scatter(df['avg'], df['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')

ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

ax[0,0].scatter(df['avg'], df['strike_rate'])
ax[0,1].scatter(df['avg'], df['runs'])
ax[1,0].hist(df['avg'])
ax[1,1].hist(df['runs'])
```

#### 3D Surface Plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10, 10, 100)

xx, yy = np.meshgrid(x,y)
z = xx**2 + yy**2

fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx, yy, z, cmap='jet')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx, yy, z, cmap='jet')
fig.colorbar(p)
```

#### Contour Plots

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contour(xx, yy, z, cmap='jet')
fig.colorbar(p)
```

```python
z = xx**2 + yy**2
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx, yy, z, cmap='jet')
fig.colorbar(p)
```
